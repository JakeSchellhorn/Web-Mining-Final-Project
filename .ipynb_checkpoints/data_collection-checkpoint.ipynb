```python
import statsapi
```

```python
battingavg_leaders21 = statsapi.league_leader_data('battingAverage', limit=5, statGroup='hitting', season = '2021')
battingavg_leaders22 = statsapi.league_leader_data('battingAverage', limit=5, statGroup='hitting', season = '2022')
battingavg_leaders23 = statsapi.league_leader_data('battingAverage', limit=5, statGroup='hitting', season = '2023')
battingavg_leaders24 = statsapi.league_leader_data('battingAverage', limit=5, statGroup='hitting', season = '2024')
```

```python
batting_avg21 = []
for i in battingavg_leaders21:
    batting_avg21.append(int(float(i[3])*1000))
batting_avg22 = []
for i in battingavg_leaders22:
    batting_avg22.append(int(float(i[3])*1000))
batting_avg23 = []
for i in battingavg_leaders23:
    batting_avg23.append(int(float(i[3])*1000))
    batting_avg24 = []
for i in battingavg_leaders24:
    batting_avg24.append(int(float(i[3])*1000))
#2 people are tied for 5th place so it wouldn't limit it to 5
```

```python
import matplotlib.pyplot as plt
bar_locations = [1,2,3,4,5]
plt.bar(list(map(lambda x: x-0.5, bar_locations)), batting_avg21, width=0.2,
         align='edge', color='g', label='2021')
plt.bar(list(map(lambda x: x-0.3, bar_locations)), batting_avg22, width=0.2,
         align='edge', color='m', label='2022')
plt.bar(bar_locations, batting_avg23, width=-0.2,
         align='center', color='c', label='2023')
plt.bar(list(map(lambda x: x+0.3, bar_locations)), batting_avg24, width=-0.2,
         align='edge', color='y', label='2024')
plt.ylabel('Batting Average')
plt.xlabel('Top 5 Batters')
plt.title('Batting Average Leaders')
plt.legend(frameon=False, loc = 'upper right', ncols = 3, fontsize = 'small', columnspacing = 0.5)
```

### This is going to get rid of junk information at the begining of the dictionary like the copyright and id

```python
pace_dict21 = statsapi.game_pace_data(season = 2021)['sports'][0]
pace_dict22 = statsapi.game_pace_data(season = 2022)['sports'][0]
pace_dict23 = statsapi.game_pace_data(season = 2023)['sports'][0]
pace_dict24 = statsapi.game_pace_data(season = 2024)['sports'][0]
```

```python
def convert_hms(hms):
    h,m,s = hms.split(':')
    return int(h)*360+int(m)*60+int(s)

convert_hms(pace_dict22['timePerHit'])
```

```python
batting_times21 = list(map(convert_hms,[pace_dict21['timePerHit'], pace_dict21['timePerRun'], pace_dict21['timePerPlateAppearance']]))
batting_times22 = list(map(convert_hms,[pace_dict22['timePerHit'], pace_dict22['timePerRun'], pace_dict22['timePerPlateAppearance']]))
batting_times23 = list(map(convert_hms,[pace_dict23['timePerHit'], pace_dict23['timePerRun'], pace_dict23['timePerPlateAppearance']]))
batting_times24 = list(map(convert_hms,[pace_dict24['timePerHit'], pace_dict24['timePerRun'], pace_dict24['timePerPlateAppearance']]))
types = ['Time per Hit', 'Time per Run', 'Time per Plate Appearance']
```

```python
import matplotlib.pyplot as plt
bar_locations = [0,1,2]
plt.barh(list(map(lambda x: x+0.5, bar_locations)), batting_times21, height=-0.2,
         align='edge', color='g', label='2022')
plt.barh(list(map(lambda x: x+0.3, bar_locations)), batting_times22, height=-0.2,
         align='edge', color='m', label='2022')
plt.barh(bar_locations, batting_times23, height=-0.2,
         align='center', color='c', label='2023')
plt.barh(list(map(lambda x: x-0.3, bar_locations)), batting_times24, height=0.2,
         align='edge', color='y', label='2024')
plt.yticks(range(len(types)), types)
plt.xlabel('Time in seconds')
plt.title('Time per Game Events')
plt.legend(frameon=False)
```
